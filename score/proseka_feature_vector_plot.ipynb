```python
import json
from pprint import pprint


file_path = f"./data/_dens_cnt_fv.json"

with open(file_path) as f:
  data = json.load(f)

pprint(data[:5])
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np


save_dir_name = "_vis-result"
dir_path = f"./data/{save_dir_name}/"
try:
    os.mkdir(dir_path)
except:
    pass

from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=0)
data_pca = pca.fit_transform(data)

plt.scatter(data_pca[:,0],data_pca[:,1])
plt.savefig(dir_path + "data-pca.png")
plt.show()


from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0)
data_tsne = tsne.fit_transform(np.array(data))

plt.scatter(data_tsne[:,0],data_tsne[:,1])
plt.savefig(dir_path + "data-tsne.png")
plt.show()

from umap import UMAP

umap = UMAP(n_components=2, random_state=0)
data_umap = umap.fit_transform(data)

plt.scatter(data_umap[:,0],data_umap[:,1])
plt.savefig(dir_path + "data-umap.png")
plt.show()
```
