```python
import numpy as np
import matplotlib.pyplot as plt
import json
from pprint import pprint
```

```python
file_path = "/content/drive/MyDrive/zemi_4/zemi/proseka/data/_feature_vector.json"

with open(file_path) as f:
  data = json.load(f)

pprint(data[:5])
```

```python

dir = "/content/drive/MyDrive/zemi_4/zemi/proseka/data/vis-result/"

from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=0)
data_pca = pca.fit_transform(data)

plt.scatter(data_pca[:,0],data_pca[:,1])
plt.savefig(dir + "data-pca.png")
plt.show()


from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0)
data_tsne = tsne.fit_transform(np.array(data))

plt.scatter(data_tsne[:,0],data_tsne[:,1])
plt.savefig(dir + "data-tsne.png")
plt.show()

!pip install umap-learn
import umap
from scipy.sparse.csgraph import connected_components

umap = umap.UMAP(n_components=2, random_state=0)
data_umap = umap.fit_transform(data)

plt.scatter(data_umap[:,0],data_umap[:,1])
plt.savefig(dir + "data-umap.png")
plt.show()
```
